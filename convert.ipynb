```python
import pandas as pd
from helpers import *

source_file="./Ciberincidentes relevantes en Argentina.xlsx"
csv_file_output="./eventos_procesados.csv"
# lee el archivo excel
excel_file=pd.read_excel(source_file,"Events")
# lo convierte al csv
excel_file.to_csv(csv_file_output,
                 index_label="id",
                 header=True
                 )
# muestra que el archivo haya sido convertido correctamente
df=pd.DataFrame(pd.read_csv(csv_file_output))

# muestra los datos
df
```

```python
#Parsea el archivo agregando datos al dataframe
# crea una copia
df2=df.copy()
# renombra las columnas
df2.rename(columns={"Group ID":'grupo_id'},inplace=True)
df2.rename(columns={"Name":'evento'},inplace=True)
df2.rename(columns={"Description":'descripcion'},inplace=True)
df2.rename(columns={"Time":'fecha'},inplace=True)
#  setea a todos los valores en 0
df2["grupo_id"]=0

# Limpia la fecha
def formatear_fecha(x):
    x2=x.split(" ")
    new_time=x2[0]+"-"+x2[1]+"-"+x2[2]
    return new_time

# Limpia la fecha
def get_anio(x):
    x2=x.split("-")
    return int(x2[0])

# Limpia la fecha
def get_mes(x):
    x2=x.split("-")
    return x2[1]

df2["fecha"]=df2["fecha"].apply(formatear_fecha)
# separa el año
df2["anio"]=df2["fecha"].copy()
df2["anio"]=df2["anio"].apply(get_anio)

# separa el mes
df2["mes"]=df2["fecha"].copy()
df2["mes"]=df2["mes"].apply(get_mes)

# limpia el nombre del evento
def get_evento_name(x):
    x2=x.strip()
    return x2[3:]

df2["evento"]=df2["evento"].apply(get_evento_name)


df_eventos=df2.copy()
df_eventos["tag"]=df_eventos["evento"].copy()
df_eventos["tag"]=df_eventos["tag"].apply(get_hashes)

# save file
df_eventos.to_csv(csv_file_output,
                  index=True,
                  header=True,
                  )
```
