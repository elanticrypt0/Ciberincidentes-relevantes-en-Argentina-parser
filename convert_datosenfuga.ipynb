```python
import pandas as pd
from helpers import *

source_file="./DatosEnFuga_Dataset_Ciberincidentes - Ciberincidentes.csv"
csv_file_output="./eventos_procesados_datosenfuga.csv"
# muestra que el archivo haya sido convertido correctamente
df_datosfuga=pd.DataFrame(pd.read_csv(source_file))

# muestra los datos
df_datosfuga
```

```python

#Parsea el archivo agregando datos al dataframe
# crea una copia
df_eventos2=df_datosfuga.copy()
# renombra las columnas
df_eventos2.rename(columns={"Organismo":'organismo'},inplace=True)
df_eventos2.rename(columns={"Mes/Año":'mes_anio'},inplace=True)
df_eventos2.rename(columns={"Tipo de organismo":'organismo_tipo'},inplace=True)
df_eventos2.rename(columns={"Tipo de ataque":'tag'},inplace=True)
df_eventos2.rename(columns={"Descripcion":'descripcion'},inplace=True)
df_eventos2.rename(columns={"Nota":'link'},inplace=True)
df_eventos2.rename(columns={"Ciudad":'ciudad'},inplace=True)
df_eventos2.rename(columns={"Pedido de acceso":'pedido_acceso'},inplace=True)
df_eventos2.rename(columns={"Fecha de incidente (mes/dia/año)":'fecha'},inplace=True)


# Limpia la fecha
def formatear_fecha(x):
    x2=x.split("/")
    new_time=x2[2]+"-"+agregar_cero_mes(x2[1])+"-"+x2[0]
    return new_time

# Limpia la fecha
def get_anio(x):
    x2=x.split(" ")
    return int(x2[1])

# Limpia la fecha
def get_mes(x):
    x2=x.split(" ")
    return x2[0].strip()

# hace que los tags sean una lista.
def convert_tag(x):
    x2=[x]
    x2 = [elemento.lower() for elemento in x2]
    return x2

df_eventos2

df_eventos2["id"]=0

df_eventos2["fecha"]=df_eventos2["fecha"].apply(formatear_fecha)
# separa el año
df_eventos2["anio"]=df_eventos2["mes_anio"].copy()
df_eventos2["anio"]=df_eventos2["anio"].apply(get_anio)

# # separa el mes
df_eventos2["mes"]=df_eventos2["mes_anio"].copy()
df_eventos2["mes"]=df_eventos2["mes"].apply(get_mes)

# normaliza el nombre del evento
df_eventos2["evento"]=df_eventos2["tag"]+" "+df_eventos2["organismo"]

# normaliza el tag
df_eventos2["tag"]=df_eventos2["tag"].apply(convert_tag)

# save file
df_eventos2.to_csv(csv_file_output,
                  index=True,
                  header=True,
                  )

df_eventos2



# df_eventos3=normalize_df(df_eventos2)

# df_eventos3



# df_eventos2["evento"]=df_eventos2["evento"].apply(get_evento_name)


# df_eventos=df_eventos2.copy()
# df_eventos["tag"]=df_eventos["evento"].copy()
# df_eventos["tag"]=df_eventos["tag"].apply(get_hashes)

# # save file
# df_eventos.to_csv(csv_file_output,
#                   index=True,
#                   header=True,
#                   )
```
