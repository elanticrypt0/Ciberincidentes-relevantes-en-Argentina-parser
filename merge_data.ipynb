```python
import pandas as pd

def combinar_datasets(df1, df2, columna_fecha):
    """
    Combina dos DataFrames basándose en una columna de fecha.
    
    Args:
    - df1: El primer DataFrame (prioritario).
    - df2: El segundo DataFrame (información adicional).
    - columna_fecha: El nombre de la columna de fecha para la combinación.
    
    Returns:
    - Un nuevo DataFrame combinado.
    """
    # Convertimos la columna de fecha a datetime para una comparación precisa
    df1[columna_fecha] = pd.to_datetime(df1[columna_fecha])
    df2[columna_fecha] = pd.to_datetime(df2[columna_fecha])
    
    # Realizamos la combinación
    df_combined = pd.concat([df1, df2]).drop_duplicates(subset=columna_fecha, keep='first')

    return df_combined

# Ejemplo de uso:
# Creamos DataFrames de ejemplo
df1 = pd.DataFrame({
    'fecha': ['2024-01-01', '2024-01-02'],
    'valor': [10, 20]
})

df2 = pd.DataFrame({
    'fecha': ['2024-01-01', '2024-01-03'],
    'valor': [15, 30]
})

# Llamamos a la función
df_combinado = combinar_datasets(df_timeline, df_datosenfuga, 'fecha')

print(df_combinado)

df_combinado.to_csv(csv_file_output,
                  index=True,
                  header=True,
                  )
```
